##1.What is the distribution of accident severity levels?

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Distribution of severity levels
plt.figure(figsize=(6,4))
sns.countplot(x="Severity", data=df, palette="viridis")
plt.title("Distribution of Accident Severity Levels")
plt.xlabel("Severity Level")
plt.ylabel("Number of Accidents")
plt.show()
```

##2.How is the accident frequency distributed by hour of the day?

```python
# Convert Start_Time properly
df["Start_Time"] = pd.to_datetime(df["Start_Time"], errors="coerce", format="mixed")

# Extract hour for accident frequency by hour
df["Hour"] = df["Start_Time"].dt.hour
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Hour", data=df, palette="magma")
plt.title("Accidents by Hour of the Day")
plt.xlabel("Hour of Day (0–23)")
plt.ylabel("Number of Accidents")
plt.show()
```

##3.Which days of the week have the highest number of accidents?

```python
# Extract day of week (0=Monday, 6=Sunday)
df["Day_of_Week"] = df["Start_Time"].dt.day_name()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Day_of_Week", data=df,
              order=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"],
              palette="cubehelix")
plt.title("Accidents by Day of the Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Accidents")
plt.show()
```

##4.What are the most common weather conditions during accidents?

```python
plt.figure(figsize=(10,6))
df["Weather_Condition"].value_counts().head(10).plot(kind="bar", color="teal")
plt.title("Top 10 Most Common Weather Conditions During Accidents")
plt.xlabel("Weather Condition")
plt.ylabel("Number of Accidents")
plt.show()
```

##5.How to identify columns with missing data and their missing percentage?


```python
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_data = missing_percent[missing_percent > 0].sort_values(ascending=False)
print(missing_data)
```

##6.How to impute missing numerical values with median?

```python
from sklearn.impute import SimpleImputer
# Example: Impute Temperature(F)
num_imputer = SimpleImputer(strategy="median")
df["Temperature(F)"] = num_imputer.fit_transform(df[["Temperature(F)"]])
```

##7.How to impute missing categorical values with the mode?


```python
from sklearn.impute import SimpleImputer
cat_imputer = SimpleImputer(strategy="most_frequent")
df[["Weather_Condition"]] = cat_imputer.fit_transform(df[["Weather_Condition"]])
```

##8.How to label encode a categorical column?


```python
from sklearn.preprocessing import LabelEncoder
# Example: Encoding 'Weather_Condition'
le = LabelEncoder()
df["Weather_Condition_Label"] = le.fit_transform(df["Weather_Condition"])
```

##9.How to one-hot encode a categorical column?


```python
# Example: One-hot encode 'Weather_Condition'
df_encoded = pd.get_dummies(df, columns=["Weather_Condition"], drop_first=True)
df_encoded.head()
```
